# Question 2 -IMDB top 100 Movies

```python
p1=requests.get("https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating")
p1
```

```python
bs=BeautifulSoup(p1.content)
```

```python
movie_name=[]
```

```python
movie_data=bs.findAll('div',class_='lister-item mode-advanced')
```

```python
# Scraping out 100 world wide movie name
for i in movie_data:
    name = i.h3.a.text
    movie_name.append(name)
```

```python
import pandas as pd
import numpy as n
```

```python
# Scraping out movie release year
year=[]

release_year=bs.findAll('span',class_='lister-item-year text-muted unbold')

for i in release_year:
    year.append(i.text.replace('(','').replace(')',''))
    

    
```

```python
# Scraping out Rating of the movie
Rating=[]

rating_movie=bs.findAll('div',class_='inline-block ratings-imdb-rating')

for i in rating_movie:
    Rating.append(i.text.replace('\n\n','').replace('\n',''))
    
```

```python
# Creating a table by using Pandas Data Frame 
movie_df=pd.DataFrame({'Name of Movie':movie_name,'Release Year':year,'Movie rating':Rating})
```

```python
movie_df
```

# Question 3- IMDB top 100 Bollywood Movies

```python
ps=requests.get('https://www.imdb.com/list/ls009997493/')
ps
```

```python
Bs=BeautifulSoup(ps.content)
```

```python
# Scrapping out Bollywood Movie Names
Movie=[]
Mo_name=Bs.findAll('h3',class_='lister-item-header')

for i in Mo_name:
    Movie.append(i.a.text.replace('\n\n','').replace('\n',''))
    
```

```python
# Scrapping out the release year of the movie
Year=[]

rel_year=Bs.findAll('span',class_='lister-item-year text-muted unbold')

for i in rel_year:
    Year.append(i.text.replace('(','').replace(')',''))
    
```

```python
#Scrapping out the Rating of the movies
rating=[]

Rat_movie=Bs.findAll('div',class_='ipl-rating-star small')

for i in Rat_movie:
    rating.append(i.text.replace('\n\n','').replace('\n',''))
    
```

```python
# Created table of top 100 bollywood movies on IMDB
Boll_mov=pd.DataFrame({'Movie':Movie,'release year':Year,'rating':rating})
Boll_mov
```

# Question 4 - List of Indian Presidents

```python
a=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
a
```

```python
b=BeautifulSoup(a.content)
```

```python
# Scrapping out Indian Presidents Name
Name_president=[]



for i in b.find_all("div",class_="presidentListing"):
    Name_president.append(i.find("h3").text.split('(')[0])
    
```

```python
# Scrapping out the Office terms of the presidents 
Term =[]

office_term=b.find_all("div",class_="presidentListing")

for i in office_term:
    Term.append(i.find('p').text.replace('Term of Office:',''))
    
```

```python
# Creating table for Indian President data 
Indian_president=pd.DataFrame({'Name of President':Name_president,'Term of Office':Term})
Indian_president
```

# Question-9 - Dineout

```python
page7=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page7
```

```python
soup7=BeautifulSoup(page7.content)
```

```python
# Creating the list of hotel name
title=[]
for i in soup7.find_all('a',class_='restnt-name ellipsis'):
    title.append(i.text)
```

```python
# Creating the list of hotel place
place=[]
for i in soup7.find_all('div',class_='restnt-loc ellipsis'):
    place.append(i.text)
```

```python
# Creating the list of hotel price
price=[]
for i in soup7.find_all('span',class_='double-line-ellipsis'):
    price.append(i.text.split('|')[0])
    
```

```python
# Creating the list of hotel cusin
Cusin=[]
for i in soup7.find_all('span',class_='double-line-ellipsis'):
    Cusin.append(i.text.split('|')[1])
    
```

```python
# Creating the list of hotel rateing 
Rating=[]
for i in soup7.find_all('div',class_='restnt-rating rating-4'):
    Rating.append(i.text)
    
```

```python
image=[]
for i in soup7.find_all('img',class_='no-img'):
    image.append(i.get('data-src'))  # scraping the url 
```

```python
Hotel_detail=pd.DataFrame({'Name':title,'Location':place,'Price':price,'Cusin':Cusin,'Rating':Rating,'URL':image})
Hotel_detail
```
